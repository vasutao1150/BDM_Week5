# 4 . Remove Values

```python
df_clean = df
df_clean = df_clean.drop('Cabin',1)
df_clean = df_clean.drop('Ticket Number',1)
df_clean.columns
```

ถ้า Column นี้มีค่าว่างอย่างน้อย 1 Column ให้ Drop Column นั้น

```python
df_clean = df_clean.dropna(1,'any')
```

ถ้า Row นี้มีค่าว่างอย่างน้อย 1 Row ให้ Drop Row นั้น

```python
df_clean = df_clean.dropna(0,'any')
```

```python
df_clean.isnull().any()
```

unique คือ การ Select Distinct

```python
df_clean['Sex'].unique()
```

```python
df_clean['Sex'] = df_clean['Sex'].replace('Female','F')
df_clean['Sex'] = df_clean['Sex'].replace('Male','M')
df_clean
```

# 6 . Replace String Values to Unique Integers

```python
df_clean['Passenger Class'].unique()
```

Factorize ---> เปลี่ยนข้อมูลที่เป็น String ให้เป็นตัวเลข

```python
p_class, levels = pd.factorize(df_clean['Passenger Class'])
```

```python
p_class
```

```python
levels
```

```python
df_clean["Passenger Class"] = p_class
df_clean
```

```python
df_clean.dtypes
```

### output ของ Factorize จะออกมาเป็น [0] แถว , [1] คอลัม 

```python
df_clean['Sex'] = pd.factorize(df_clean['Sex'])[0]
df_clean['Survived'] = pd.factorize(df_clean['Survived'])[0]
df_clean.dtypes
```

# 7 . Set index
